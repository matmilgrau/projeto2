#MANUAL DO JOGO

#####OBJETIVO DO JOGO:
- Você controla o personagem (𖠋) em uma grade cheia de obstáculos (#). Seu objetivo é navegar livremente pelos “.” ao longo do mapa e alcançar o destino final (□) sem sair dos limites do mapa.

#####SÍMBOLOS DO JOGO:
- "𖠋"  =  Seu personagem
- "#"  =  Obstáculos. Não é possível passar por eles.
- "□"  =  Objetivo principal do nível 1. Chegue até este portal para avançar de nível.
- "❚"  = Porta que volta o player do nível 2 para o nível 1
- "?" = Objetivo principal do nível 2. Chegue até esse portal para avançar de nível
- "$" = Objetivo principal do nível 3. Chegue até esse portal para vencer o jogo
- "."  =  Espaço vazio. O jogador pode se mover livremente.
- "𓁍" = Inimigo. Ao tocá-lo, você perde.

#####CONTROLES:
- A cada comando inserido (em letras minúsculas), pressione ENTER para confirmar a ação.
- W : Cima
- A : Esquerda
- S : Baixo
- D : Direita
- Retornar: O jogo salva seu progresso ao mudar de fase, e ao usar o comando "retornar", o programa retorna para última fase que você esteve

#####NÍVEIS:
- Cada nível apresenta um cenário diferente, quanto maior o nível, maior o tamanho do mapa e mais complexo o caminho.
- Nível 1 - Quintal : Fácil
- Nível 2 - Casa : Médio
- Nível 3 - Labirinto: Difícil

#####REGRAS E CONDIÇÕES:
- Caso o personagem tente se mover para o obstáculo (#), o jogo irá exibir “Obstáculo!”, porém você continuará jogando normalmente.
- Se tentar se mover para fora do limite do mapa, o jogo irá exibir “Você saiu da grade!” e encerrará o nível com GAME OVER!
- Ao mover o personagem (𖠋) para a posição do portal (□), o jogo o levará para o próximo nível.
- Ao mover o personagem (𖠋) para a posição do portal (?), o jogo o levará para o próximo nível.
- Ao mover o personagem (𖠋) para a posição do portal ($), o jogo voltará para o menu principal exibindo uma tela de vencedor.
- Ao mover o personagem (𖠋) para a posição da porta (❚), o jogo o levará de volta para o nível anterior.


#####MENU PRINCIPAL:
- Ao iniciar o jogo, você verá o menu principal com as opções:
- [1] START : Abre uma página para a seleção de nível de preferência ([1], [2] ou [3]) ou para voltar ao menu principal escolhendo o número [0].
- [2] SAIR DO JOGO : Encerra o programa.

#COMO CONFIGURAR UMA FASE:
- Usamos os comandos de funções para poder executar sempre sem ter que escrever tudo, já que há momentos em que você vai de uma fase para outra, o que é facilitado com o encurtamento de comandos. Para criar as fases, usamos uma fórmula que determina as dimensões da fase, em que o número do nível é multiplicado por 4 e depois tudo isso é somado por 6 ====> (nível*4)+6 =====> e de resto, utilizamos os "." para gerar o mapa que pode ser 10x10, 14x14 ou 18x18, o qual tudo depende do nível que você escolhe.
- Utilizamos alguns comandos recomendados pelo professor para facilitar a criação do espaço em que é o mapa.
- Também aplicamos o comando **while** para que, em todo momento, o comando se repitisse para gerar as fases automaticamente.


```python
from IPython.display import clear_output
import random

nivel=0

# Gera uma grade quadrada de pontos com tamanho baseado no nível atual
def montar_grade(tamanho= nivel*4 + 6):
    linha = ['.']*tamanho
    grade = []
    for i in range(tamanho):
        grade.append(linha.copy()) # .copy() evita que todas as linhas apontem para o mesmo objeto
    return grade

def exibir_grade(grade):
    for lin in grade:
        print(' '.join(lin))

def jogar(nivel): # Criamos a função chamada "jogar", ou seja, jogar = nivel
    tamanho = nivel*4 + 6
    grade = montar_grade(tamanho)

    if nivel == 2:
        obstaculos = [(11,1),(11,2),(11,4),(11,5),(10,1),(10,2),(10,3),(10,4),
         (10,5),(9,1),(9,2),(9,3),(9,4),(9,5),(8,1),(8,2),(8,4),(8,5),(7,2),(7,3),
          (7,4),(6,3),(1,1),(1,6),(2,1),(3,2),(5,0),(5,1),(5,5),(5,6),(2,4),(3,4),
           (4,4),(1,8),(2,6),(2,8),(3,6),(3,7),(3,8),(3,9),(1,10),(1,11),(3,11),
            (4,11),(4,12),(5,13),(7,8),(8,8),(8,9),(8,10),(8,12),(11,10),(12,7),(12,8),(12,11)]
        inimigos = [(5,8),(11,7)]
        pos_inicial = (12, 3)
        grade[11][3]="❚" # Entrada do labirinto (portal para o nível 1)
        grade[2][7]="?"
        grade[8][3]="♰" # Símbolo decorativo/cenário
        print("=== NÍVEL 2 ===\nObjetivo: procure e entre no labirinto")
    elif nivel == 1:
        obstaculos = [(0,0),(1,0),(2,0),(2,3),(3,0),(3,2),(4,0),(4,2),(4,8),(4,9),
         (5,9),(5,0),(6,0),(7,0),(8,0),(9,0),(0,1),(7,1),(9,1),(0,2),(9,2),(0,3),
          (3,3),(7,3),(8,3),(9,3),(0,4),(2,4),(7,4),(0,5),(7,5),(9,5),(0,6),
           (5,6),(9,6),(0,7),(3,7),(4,7),(5,7),(6,7),(9,7),(0,8),(9,8),(0,9),(1,9),
            (2,9),(3,9),(6,9),(7,9),(8,9),(9,9)]
        inimigos = [(4,4),(6,2)]
        pos_inicial = (1,4)
        grade[9][4]="□" # Saída da casa (porta)
        print("=== NÍVEL 1 ===\nObjetivo: saia da casa")
    elif nivel == 3:
        obstaculos = [(0,3),(1,0),(1,1),(1,3),(1,5),(1,6),(1,7),(1,8),(1,9),(1,10),
         (1,11),(1,12),(1,13),(1,14),(1,15),(2,1),(2,3),(2,5),(2,15),(3,1),(3,5),(3,7),
          (3,8),(3,9),(3,10),(3,11),(3,12),(3,15),(4,1),(4,2),(4,3),(4,5),(4,7),(4,12),
           (4,14),(4,15),(5,3),(5,7),(5,9),(5,10),(5,12),(6,0),(6,1),(6,3),(6,4),(6,5),
            (6,7),(6,9),(6,12),(6,14),(6,15),(6,16),(7,0),(7,5),(7,6),(7,7),(7,9),(7,10),
             (7,11),(7,12),(7,16),(8,0),(8,2),(8,3),(8,14),(8,16),(9,0),(9,3),(9,5),(9,6),
              (9,7),(9,8),(9,9),(9,10),(9,11),(9,12),(9,14),(9,16),(10,0),(10,14),(10,16),
               (11,0),(11,2),(11,3),(11,4),(11,5),(11,6),(11,7),(11,8),(11,9),(11,10),(11,12),
                (11,14),(11,16),(12,0),(12,10),(12,12),(12,14),(12,16),(13,0),(13,2),(13,3),
                 (13,4),(13,5),(13,6),(13,7),(13,10),(13,12),(13,14),(13,15),(13,16),(14,7),
                  (14,9),(14,10),(14,12),(14,14),(14,16),(15,1),(15,2),(15,3),(15,4),(15,5),
                   (15,7),(15,12),(15,13),(15,14),(15,16),(16,1),(16,5),(16,7),(16,8),(16,9),
                    (16,10),(16,14),(16,16),(16,17),(17,1),(17,10)]
        inimigos = []
        grade[17][17]="$" # Objetivo final: encontrar o "$" vence o jogo
        pos_inicial = (0, 0)
        print("=== NÍVEL 3 ===\nObjetivo: encontre a saída ($) do labirinto")

 # Marca todos os obstáculos na grade com '#'
    for (obs_i, obs_j) in obstaculos:
        grade[obs_i][obs_j] = '#'
    for (obs_i,obs_j) in inimigos:
        grade[obs_i][obs_j] = "𓁍"
# Posiciona o jogador '𖠋' na posição inicial do nível
    i, j = pos_inicial
    grade[i][j] = '𖠋'
    exibir_grade(grade)

    status_ok = True
    carlos=1
    while status_ok:
        cmd = input('\nComando (w/a/s/d/retornar): ') # Inserir comando desejado
        clear_output() # Limpar a saída de texto

        if cmd == 'retornar':
            print("Saindo do nível...")
            status_ok = False
            break # Para o programa ao inserir a palavra "retornar" na entrada de texto

# Calcula a nova posição
        nova_linha = i
        nova_coluna = j

        if cmd == 'w':
            nova_linha -= 1 # Move para cima
        elif cmd == 's':
            nova_linha += 1 # Move para baixo
        elif cmd == 'a':
            nova_coluna -= 1 # Move para esquerda
        elif cmd == 'd':
            nova_coluna += 1 # Move para direita

 # Verifica se a nova posição está dentro dos limites da grade
        if 0 <= nova_linha < tamanho and 0 <= nova_coluna < tamanho:
            if grade[nova_linha][nova_coluna] == '#':
                print("Obstáculo!")
            elif grade[nova_linha][nova_coluna] == "𓁍":
                status_ok = False
                print(f"VOCÊ MORREU!")
                break
            elif grade[nova_linha][nova_coluna] == "❚":
                nivel=1 # Caso o jogador entre no portal, irá reiniciar no nível 1 em posição específica (foi definido logo em seguida)
                jogar(nivel)
            elif grade[nova_linha][nova_coluna] == "□":
                nivel=2 # Avança para o próximo nível(2)
                jogar(nivel)
            elif grade[nova_linha][nova_coluna] == "?":
                nivel=3 # Avança para o próximo nível(3)
                jogar(nivel)
            elif grade[nova_linha][nova_coluna] == "$":
                status_ok = False # Jogador encontrou a saída final: vitória
                print("╔═══════════════════════════╗")
                print("║       VOCÊ VENCEU!!!      ║")
                print("╚═══════════════════════════╝")
            else: # Apaga a posição antiga e move o jogador
                if(nivel==1):
                  inimigo = [1,0,-1] #possíveis posições que o inimigo pode se mover
                  escolha1 = random.choice(inimigo)
                  escolha2 = random.choice(inimigo) #as duas variáveis escolhem um número aleatório usando random.choice
                  if(grade[1][4] == "𖠋"): #isso só funciona quando o boneco estiver na pos inicial
                    pos_inimigo1=4
                    pos_inimigo2=4
                  grade[pos_inimigo1][pos_inimigo2] = "." #grade antiga
                  pos_inimigo1=4+escolha1
                  pos_inimigo2=4+escolha2
                  grade[pos_inimigo1][pos_inimigo2] = "."
                  grade[pos_inimigo1][pos_inimigo2] = "𓁍" #grade nova
                grade[i][j] = '.'
                i, j = nova_linha, nova_coluna
                grade[i][j] = '𖠋'
                if(nivel==2):
                  inimigo = [1,0,-1]
                  escolha1 = random.choice(inimigo)
                  escolha2 = random.choice(inimigo)

                  if((grade[13][3] == "𖠋" or grade[12][4] == "𖠋" or grade[12][2]== "𖠋") and carlos==1):
                    pos_inimigo1=5
                    pos_inimigo2=8
                    pos_inimigo3=11
                    pos_inimigo4=7
                    carlos=2

                  grade[pos_inimigo1][pos_inimigo2] = "."
                  pos_inimigo1=5+escolha1
                  pos_inimigo2=8+escolha2
                  grade[pos_inimigo1][pos_inimigo2] = "."
                  grade[pos_inimigo1][pos_inimigo2] = "𓁍"
                  grade[pos_inimigo3][pos_inimigo4] = "."
                  pos_inimigo3=11+escolha1
                  pos_inimigo4=7+escolha2
                  grade[pos_inimigo3][pos_inimigo4] = "."
                  grade[pos_inimigo3][pos_inimigo4] = "𓁍"
                grade[i][j] = '.'
                i, j = nova_linha, nova_coluna
                grade[i][j] = '𖠋'


        else:
            print("Você saiu da grade!\nGAME OVER!")
            status_ok = False
            break
        exibir_grade(grade)

    if not status_ok:
        print("(programa encerrado)")

rodando = True # tela de início
while rodando:
    print("╔══════════════════════════╗")
    print("║        BEM VINDO(A)!     ║")
    print("║==========================║")
    print("║  [1] START               ║")
    print("║  [2] SAIR DO JOGO        ║")
    print("╚══════════════════════════╝")

    opcao = input("Escolha uma opção: ")
    clear_output()

    if opcao == '1':
        print("╔════════════════════════════════════╗")
        print("║           ESCOLHA O NÍVEL          ║")
        print("║====================================║")
        print("║  [1] NÍVEL 1 - Casa (FÁCIL)        ║")
        print("║  [2] NÍVEL 2 - Quintal (MÉDIO)     ║")
        print("║  [3] NÍVEL 3 - Labirinto (DIFÍCIL) ║")
        print("║  [0] VOLTAR AO MENU                ║")
        print("╚════════════════════════════════════╝")

        nivel_escolhido = input("Escolha o nível: ")
        clear_output() # Limpar a saída de texto

        if nivel_escolhido == '1' or nivel_escolhido == '2' or nivel_escolhido == '3':
          jogar(int(nivel_escolhido))
        elif nivel_escolhido == '0':
            pass  # Volta ao menu principal
        else:
            print("Opção inválida!")

    elif opcao == '2': # Encerra o jogo
        print("SAINDO DO JOGO...\nJOGO ENCERRADO!")
        rodando = False
    else: # Caso insira um comando não identificado
        print("Opção inválida! Tente novamente.")
```
